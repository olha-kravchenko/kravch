## Data Visualization - Iris

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as dss
```

### Load the Data

```python
ds = dss.load_iris()
```

```python
feature_names = ds['feature_names']
feature_names
```

```python
ds['target']
```

```python
# create data from the first 5 columns
df = pd.DataFrame(
    data = ds['data'],
    columns = feature_names
)

df.info()
```

```python
# add a column
df['target'] = ds['target']
df.info()
```

### Pairwise correlation of columns

```python
corr = df.corr()
fig, ax = plt.subplots(figsize=(5, 4)) # size of the box
sns.heatmap(corr,  cmap='coolwarm', annot=True, fmt=".2f")
plt.show()
```

### Boxplots
for more examples see https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
for i in range(len(feature_names)):
    figure = plt.figure()
    ax = sns.boxplot(x='target', y=feature_names[i], data=df)
```

### Pairplot 
for more examples see 
https://seaborn.pydata.org/generated/seaborn.pairplot.html

```python
sns.pairplot(df, hue="target", palette = 'Set1')
```

### Додаткові візуалізації

```python
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (7, 4)

df = sns.load_dataset("iris")
```

#### 1) Класичний pairplot з підсвіткою видів


```python
pp = sns.pairplot(df, hue="species", corner=True, diag_kind="kde", plot_kws=dict(alpha=0.85, s=28))
pp.fig.suptitle("Iris: Pairplot (species)", y=1.02)
plt.show()
```

#### 2) Scatterplot ключових пар ознак

```python
plt.figure()
sns.scatterplot(data=df, x="sepal_length", y="petal_length", hue="species")
plt.title("Iris: sepal_length vs petal_length")
plt.xlabel("sepal_length"); plt.ylabel("petal_length")
plt.show()

plt.figure()
sns.scatterplot(data=df, x="petal_width", y="petal_length", hue="species")
plt.title("Iris: petal_width vs petal_length")
plt.xlabel("petal_width"); plt.ylabel("petal_length")
plt.show()
```

#### 3) Розподіли ознак по видах (box/violin)

```python
for col in ["sepal_length", "sepal_width", "petal_length", "petal_width"]:
    plt.figure()
    sns.boxplot(data=df, x="species", y=col)
    plt.title(f"Iris: {col} по видах")
    plt.xlabel("species"); plt.ylabel(col)
    plt.show()
```
